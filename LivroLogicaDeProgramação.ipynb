Operadores aritmeticos

Potenciação e radiciação

```python
numero1 = 5;
numero2 = 2;
ad = numero1 + numero2;
sub = numero1 - numero2;
mult = numero1 * numero2;
div = numero1/numero2;
divInt = numero1//numero2;
mod = numero1%numero2;
print(f'Entrada dos numeros\n 1º número → {numero1}\n 2º número → {numero2}');
print(f'Adição = {ad}\nSubtração = {sub}\nMultiplição = {mult}\nDivisão = {div}\nDivisão inteira = {divInt}\nMod = {mod}\n') ;
```

```python
import math;
num1 = 9;
num2 = 2;
pot = pow(num1,num2);
raiz = math.sqrt(pot);
print(f'Entrada dos numeros\n 1º número → {num1}\n 2º número → {num2}');
print(f'Potencia = {pot}\nRaiz = {raiz}');
```

```python
#entrada de dados
h = float(input("Digite a altura: "))
r = float(input("Digite o raio: "))
#constantes
pi = 3.14
precoLitro = 50.00
#processamento
area = (pi * pow(r,2))+(2*pi*r*h)
litro = area/3
qde = litro/5
custo = qde * precoLitro
#saida
print('Dados ')
print(f'Area: {area}\nLitros necessários: {litro}\nLatas de necessáriass: {qde}\nCusto: R${custo:.2f}')
```
