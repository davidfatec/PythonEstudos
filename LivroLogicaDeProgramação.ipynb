```python
import math;
num1 = 9;
num2 = 2;
pot = pow(num1,num2);
raiz = math.sqrt(pot);
print(f'Entrada dos numeros\n 1º número → {num1}\n 2º número → {num2}');
print(f'Potencia = {pot}\nRaiz = {raiz}');
```

```python
#entrada de dados
h = float(input("Digite a altura: "))
r = float(input("Digite o raio: "))
#constantes
pi = 3.14
precoLitro = 50.00
#processamento
area = (pi * pow(r,2))+(2*pi*r*h)
litro = area/3
qde = litro/5
custo = qde * precoLitro
#saida
print('Dados ')
print(f'Area: {area}\nLitros necessários: {litro}\nLatas de necessáriass: {qde}\nCusto: R${custo:.2f}')
```

```python
#entrada
n1 = int(input("Digite o primeiro numero: "))
n2 = int(input("Digite o segundo numero: "))
operador = str(input("Digite o operador matemático: ")).strip()[0]
#variaveis de apoio 
tipoOperacao = ""
operacao = 0
#processamento
if operador == "+":
  tipoOperacao = 'Soma'
  operacao = n1 + n2
elif operador == "-":
  tipoOperacao = 'Subtração'
  operacao = n1 - n2
elif operador == "*":
  tipoOperacao = 'Multiplicação'
  operacao = n1 * n2 
elif operador == "/":
  tipoOperacao = 'Divisão'
  operacao = n1 / n2
else:
  tipoOperacao = 'entrada do operador é inválida, não será feito operação'
  operador = 'e'
  operacao = "Erro na entrada do operador" 
#saida
print(f'A {tipoOperacao} de {n1} {operador} {n2} = {operacao} ')
```

```python
#entrada
branco = 0
tinto = 0
rose = 0
total = 0
#processamento
entrada = ''
while (entrada != 'F'):
   entrada = str(input('Digite o tipo de vinho Branco → B Tinto → T Rose → R\n →')).upper()[0]
   if(entrada != "B" and entrada != "T" and entrada != "R" and entrada != "F"):
      print('Entrada inválida')
   else:
     if entrada == "B":
       total = total + 1
       branco = branco + 1
     elif entrada == "T":
       total = total + 1
       tinto = tinto + 1  
     elif entrada == "R":
       total = total + 1
       rose = rose + 1
print(f'Total de vinhos registrados {total}')
print(f'Branco {branco} - {branco/total*100:.2f}% ')
print(f'Tinto {tinto} - {tinto/total*100:.2f}% ')
print(f'Rose {rose} - {rose/total*100:.2f}% ')
```
